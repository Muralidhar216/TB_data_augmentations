```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("CODA_TB_Solicited_Meta_Info.csv")
df.head(10)
```

```python
list=[]
for i,row in df.iterrows():
    if (row["tb_status"] == 1):
        list.append(row["filename"])
print(list)
print(len(list))
```

```python
audiolist=[]
for i in list:
    output_filename = "solicited/"+i
    audiolist.append(output_filename)
print(audiolist)
print(len(audiolist))
```

```python
temp=audiolist[0]
x=temp.split("/")
print(temp)
print(x)
```

```python
import os
import librosa
import soundfile as sf
from audiomentations import Compose, AddGaussianNoise, PitchShift

#hell=[]
# Set up the augmentation pipeline
augmentations = Compose([
    AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.01, p=0.5),
    PitchShift(min_semitones=-4, max_semitones=4, p=0.5),
])

#input_dir = "solicited/"
output_dir = "audiomentations/finalaudio/"

for file_name in audiolist:
    if file_name.endswith(".wav"):
        #input_path = os.path.join(input_dir, file_name)
        #hell.append(file_name)

        x=file_name.split("/")
        output_filename = os.path.splitext(x[1])[0]+"_aug"+".wav"
        output_path = os.path.join(output_dir,output_filename )

        # Load the audio file
        audio, sample_rate = librosa.load(file_name)

        # Apply augmentations
        augmented_audio = augmentations(samples=audio, sample_rate=sample_rate)

        # Save the augmented audio to a new file
        sf.write(output_path, augmented_audio, sample_rate)
```
